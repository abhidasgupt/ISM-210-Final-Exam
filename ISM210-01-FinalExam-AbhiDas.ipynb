4.a) Return the top 3 companies in each STATE based on the average "Long-Term Debt – Total" (5 pts)
* You can use a user-defined function following the lecture notes
* You will need to first aggregate at the company level to account for the fact that a company appears multiple times due to the column YEAR.
* The unique identifier of company in the dataset is GVKEY. You can also use TICKER or Company Name

```python
# write code here
def top(df, column, n=3):
    return df.sort_values(by=column)[-n:]
FirmAndExecutivesFinancialDataAggStateCorp = FirmAndExecutivesFinancialData.groupby([FirmAndExecutivesFinancialData['STATE'], FirmAndExecutivesFinancialData['Company Name']]).mean()
FirmAndExecutivesFinancialDataAggStateCorp = FirmAndExecutivesFinancialDataAggStateCorp.reset_index()

FirmAndExecutivesFinancialDataAggState = FirmAndExecutivesFinancialDataAggStateCorp.groupby(FirmAndExecutivesFinancialDataAggStateCorp['STATE'])
FirmAndExecutivesFinancialDataAggState.apply(top, column='Long-Term Debt – Total', n=3)[['Company Name','Long-Term Debt – Total']]
```

* You can use a user-defined function following the lecture notes
* GENDER contains the gender of CEOs
* The unique identifier of company in the dataset is GVKEY. You can also use TICKER or Company Name

```python
# write code here
def top(df, column, n=3):
    return df.sort_values(by=column)[-n:]

FirmAndExecutivesFinancialDataMale = FirmAndExecutivesFinancialData[FirmAndExecutivesFinancialData["GENDER"] == "MALE"]

FirmAndExecutivesFinancialDataMaleAggStateCorp = FirmAndExecutivesFinancialDataMale.groupby([FirmAndExecutivesFinancialDataMale['STATE'], FirmAndExecutivesFinancialDataMale['Company Name']]).mean()
FirmAndExecutivesFinancialDataMaleAggStateCorp = FirmAndExecutivesFinancialDataMaleAggStateCorp.reset_index()

FirmAndExecutivesFinancialDataMaleAggState = FirmAndExecutivesFinancialDataMaleAggStateCorp.groupby(FirmAndExecutivesFinancialDataMaleAggStateCorp['STATE'])
FirmAndExecutivesFinancialDataMaleAggState.apply(top, column='Long-Term Debt – Total', n=3)[['Company Name','Long-Term Debt – Total']]
```

Using **patsy** and **statsmodels**, fit two models that predict "Long-Term Debt – Total" (10 pts)
For each model, choose a different combination of predictors. You can include categorical variables (e.g., State, Year, etc.), and/or numerical variables (e.g., Salary, Shares Owned, etc.)
* Each model should have at least 5 predictors
    * DO NOT USE ANY OF THE FOLLOWING VARIABLES: Net Income, Gross Profit, Long-Term Debt – Total, Assets – Total, and Debt – Due in One Year
    * **IF I NOTICE THAT YOU LITERALLY COPIED THE PREDICTORS I USED IN THE PRACTICE EXAM, I WILL GIVE YOU ZERO ON THIS QUESTION.**
* Each model is worth 5 pts
* Select the best model based on the R-Squared coefficient (5 pts)
* The best model is the one that has the highest R-Squared coefficient

```python
FirmAndExecutivesFinancialData.columns
```

```python
# write code here
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy

y, X = patsy.dmatrices('Q("Long-Term Debt – Total") ~ AGE + PENSION_CHG + Q("Shares Owned") + CITY + C(YEAR) + Q("Unexercised Exercisable Options")', FirmAndExecutivesFinancialData)
model_1 = sm.OLS(y, X)
results_model_1 = model_1.fit()
print(results_model_1.summary())
```

```python
y, X = patsy.dmatrices('Q("Long-Term Debt – Total") ~ Q("Capital Surplus") + GENDER + SALARY + Q("STOCK_AWARDS_Fair Value") + C(ZIP) + C(NAICS) + Q("INDDESC")', FirmAndExecutivesFinancialData)
model_2 = sm.OLS(y, X)
results_model_2 = model_2.fit()
print(results_model_2.summary())
```

*Based on the R-Squared Coefficient, Model 2 is the better model*
